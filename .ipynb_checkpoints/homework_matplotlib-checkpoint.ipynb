```python
# Importar de la libería

import matplotlib.pyplot as plt
import numpy as np
```

```python
# Desarrollo de la función
x = np.arange(-15,15,1)
sigma1 = 10
sigma2 = 20
sigma3 = 30

y1 = (1/(sigma1*np.sqrt(2*np.pi)))*np.exp((-1/2)*(x/sigma1)**2)
y2 = (1/(sigma2*np.sqrt(2*np.pi)))*np.exp((-1/2)*(x/sigma2)**2)
y3 = (1/(sigma3*np.sqrt(2*np.pi)))*np.exp((-1/2)*(x/sigma3)**2)

# Gráfica

plt.plot(x,y1, linestyle = "-", label = "Sigma = 10")
plt.plot(x,y2, linestyle = "-.", label = "Sigma = 20")
plt.plot(x,y3, linestyle = "--", label = "Sigma = 30")
plt.legend()
plt.title("Exploración del valor Sigma")
plt.xlabel("Variable")
plt.ylabel("Frecuencia")

plt.show()
```
