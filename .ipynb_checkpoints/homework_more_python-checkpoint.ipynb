```python
# 1. Convierte la siguiente lista de temperaturas en Fahrenheit a grados Centígrados usando un for loop.

temperaturas_f = [30.2, 41, 115.7, 77, 130.46, 59, 86, 63.5, 23]
temperatura_c = []
for farenheit in temperaturas_f:
    celcius = 5/9*(farenheit - 32)
    temperatura_c.append(celcius)
#print(temperatura_c)
temperatura_c
```

```python
# 2.Utiliza for loops y condicionales para generar una alerta para temperaturas:

for temperatura in temperatura_c:
    if temperatura >= 35:
        print(temperatura,"es alta temperatura")
    elif temperatura <= 0:
        print(temperatura,"es baja temperatura")
    else:
        print(temperatura,"es normal")
```

```python
# 3.Genera dos funciones:
# 3.1 una función que tome un único valor de temperatura en Fahrenheit y devuelve el mismo valor pero en Celsius.

def FahrenheitCelcius(temperatura):
    """
    Transforma la temperatura de Fahrenheit a Celcius
    
    Parameters
    --------------------------------
    temperatura: float or int
                 Temperatura a Fahrenheit
    
    Returns
    --------------------------------
    resultado: floar or int
               Temperatura en Celcius
    """
    
    return 5/9*(temperatura-32)
```

```python
FahrenheitCelcius(41)
```

```python
# 3.1 Para un compilado de temperaturas:

def FahrenheitCelcius(t1,t2,t3,t4):
    temperatura_f = [t1,t2,t3,t4]
    temperatura_c = []
    for temperatura in temperatura_f:
        celcius = 5/9*(temperatura - 32)
        temperatura_c.append(celcius)
    return temperatura_c

FahrenheitCelcius(30.1, 41.5, 115, 77)
```

```python
# 3.2 Una función que toma un único valor de temperatura e imprime su correspondiente alerta de temperatura

def Alerta(temperatura):
    """
    Convierte la temperatura de Fahrenheit a Celcius
    
    Parameters
    --------------------------------
    temperatura: float or int
                 Temperatura en Fahrenheit
    
    Returns
    --------------------------------
    resultado: floar or int
               Alerta de temperatura
    """
    
    celcius = 5/9*(temperatura-32)
    if celcius >= 35:
        print(celcius,"es alta temperatura")
    elif celcius <= 0:
        print(celcius,"es baja temperatura")
    else:
        print(celcius,"es normal")

Alerta(41)
```

```python
# 3.2 Alerta para un compilado de temperaturas

def Alerta(t1,t2,t3,t4):
    fahrenheit = [t1,t2,t3,t4]
    for temperatura in fahrenheit:
        celcius = 5/9*(temperatura - 32)
        if celcius >= 35:
            print(celcius,"es alta temperatura")
        elif celcius <= 0:
            print(celcius,"es baja temperatura")
        else:
            print(celcius,"es normal")

Alerta(30.1, 41.5, 115, 77)
```

```python
# Monitorear la velocidad del viento y la visibilidad del aeropuerto. Crear una función de alerta.

def Alerta_vuelo(velocidad,visibilidad):
    """
    Indicar al piloto cómo está el clima para el despege de los aviones
    
    Parameters
    --------------------------------
    velocidad: int or float
            velocidad del viento en nudos
    visibilidad: int or float
                 en metros
       
    Returns
    --------------------------------
    resultado: str
               Alerta de condición climática para el despegue
    """
    
    if velocidad > 12 or visibilidad < 100:
        return "Hay mal tiempo, la pista debe cerrarse"
    elif 8 < velocidad < 12:
        return "Precaución, hay vientos fuertes"
    elif 100 <= visibilidad <= 500:
        return "Cuidado, existen bajas condiciones de visibilidad"
    else:
        return "Buenas condiciones climáticas, que tenga un buen vuelo"    
```

```python
# Función Gaussiana
import numpy as np
import matplotlib.pyplot as plt

# Para un solo punto
def Gauss1(x, sigma=1):
    
    """
    Calcular la función Gaussiana para un solo punto en las absisas
    
    Parameters
    --------------------------------
    x: int or float
       valor de la absisa
    sigma: int or float
           valor de sigma cuyo defecto es de 1
       
    Returns
    --------------------------------
    resultado: int o float
               el valor de y (eje de las ordenadas)
    """
    
    y = (1/(sigma*np.sqrt(2*np.pi)))*np.exp((-1/2)*(x/sigma)**2)          
    return y

# Para un rango de puntos
def Gauss2(x_inicio, x_final, sigma=1):
    """
    Calcular la función Gaussiana para un rango determinado en x
    
    Parameters
    --------------------------------
    x: int or float
       valor de la absisa
    sigma: int or float
           valor de sigma cuyo defecto es de 1
       
    Returns
    --------------------------------
    resultado: plt
               valores de "y" (eje de las ordenadas) y su gráfica               
    """
    
    listaX = np.arange(x_inicio, x_final, 1)
    listaY = []
    for x in listaX:
        y = (1/(sigma*np.sqrt(2*np.pi)))*np.exp((-1/2)*(x/sigma)**2)
        listaY.append(y)
    
    plt.plot(listaX,listaY)
    plt.xlabel("variable")
    plt.ylabel("frecuencia")
    plt.title("Función Gausseana")
    plt.show()
    return "los valores del eje y son ", listaY
```

```python
print("Para x = 5 y resulta de",Gauss1(5))
```

```python
print(Gauss2(1,10,1), "Para x de 1 a 10, con sigma = 1 resulta esta gráfica")
```

```python
# Con 3 sigmas diferentes

def Gauss2(x_inicio, x_final, sigma1,sigma2,sigma3):
    """
    Calcular la función Gaussiana para un rango determinado en x, con 3 diferentes sigmas
    
    Parameters
    --------------------------------
    x: int or float
       valor de la absisa
    sigma: int or float
           valores de sigma
       
    Returns
    --------------------------------
    resultado: plt
               Las gráficas de las funciones con distintos sigmas                
    """
    
    listaX = np.arange(x_inicio, x_final, 1)
    listaG = [sigma1,sigma2,sigma3]
    listaY = []
    for sigma in listaG:
        for x in listaX:
            y = (1/(sigma*np.sqrt(2*np.pi)))*np.exp((-1/2)*(x/sigma)**2)
            listaY.append(y)
        
        plt.plot(listaX,listaY)
        plt.xlabel("variable")
        plt.ylabel("frecuencia")
        plt.title("Función Gausseana para sigma de "+str(sigma))
        plt.show()
        
        listaY = []  

Gauss2(1,10,10,20,30)
```

```python
# Bonus track

# Para u = 0
def Gauss3(x_inicio, x_final, sigma=1, u=0):
    """
    Calcular la función Gaussiana para un rango determinado en x, con sigma y mu predefinidos 
    
    Parameters
    --------------------------------
    x: int or float
       valor inicial y final de la absisa
    sigma: int or float
           valor de sigma
    mu: int or float
        valor de mu
       
    Returns
    --------------------------------
    resultado: plt
               La gráfica de las función  
    """
    
    listaX = np.arange(x_inicio, x_final, 1)
    listaY = []
    for x in listaX:
        y = (1/(sigma*np.sqrt(2*np.pi)))*np.exp((-1/2)*((x-u)/sigma)**2)
        listaY.append(y)
        
    plt.plot(listaX,listaY)
    plt.xlabel("variable")
    plt.ylabel("frecuencia")
    plt.title("Función Gausseana")
    plt.show()
    
    return "los valores del eje y son ", listaY

# Para tres valores distintos de u

def Gauss4(x_inicio, x_final, u1, u2, u3):
    """
    Calcular la función Gaussiana para un rango determinado de x con diferentes valores para mu  
    
    Parameters
    --------------------------------
    x: int or float
       valor inicial y final de la absisa
    mu: int or float
        valores de mu
       
    Returns
    --------------------------------
    resultado: plt
               Las gráficas de las funciones con distintos valores para mu  
    """
    
    listaX = np.arange(x_inicio, x_final, 1)
    listaU = [u1, u2, u3]
    listaY = []
    for u in listaU:
        for x in listaX:
            y = (1/(1*np.sqrt(2*np.pi)))*np.exp((-1/2)*((x-u)/1)**2)
            listaY.append(y)
        
        plt.plot(listaX,listaY)
        plt.xlabel("variable")
        plt.ylabel("frecuencia")
        plt.title("Función Gausseana para u de "+str(u))
        plt.show()
        
        listaY = []
```

```python
Gauss3(1,10)
Gauss4(1,10,5,10,2)
```
